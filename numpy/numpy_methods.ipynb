```python
import numpy as np
```

```python
# array() --> it's numpy method use to create array in python
'''
    syntax:
        np.array(value) --> value can be list/tuple/nested list/nested tuple
'''

arr=np.array([1,2,3,4,5])
print(arr,type(arr))
```

```python
arr=np.array([1,2,3,4,5.5])
print(arr,type(arr))
```

```python
arr=np.array([1,2,3,4,5,'sum'])
print(arr,type(arr))
arr
```

```python
arr=np.array([1,2,3,4,5.7,True])
print(arr,type(arr))
```

```python
arr=np.array([1,1,5,10,45.6],dtype = int)
arr
```

```python
arr=np.array(['hi','hello','hiiiiiiiiiiiii'],dtype='<U5')
arr
```
