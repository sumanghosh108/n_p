```python
import numpy as np
```

### np.sum()
    syntax: 
        np.sum() --> returns value
### mean()
    * np.mean(arr) --> returns value
    * np.var(arr) --> returns value
    * np.std(arr) --> returns value

```python
arr1=np.array([1,2,3,4])
arr2=np.array([[1,2,3,4],[5,6,7,8]])
print(arr1)
print(arr2)
print(arr1.ndim)
print(arr2.ndim)
```

### sum -> sumof the all elements of an array

```python
np.sum(arr1)
```

## axis = 0 -> columns wise

```python
np.sum(arr1,axis=0)
```

## axis = 1 -> row wise

```python
np.sum(arr2,axis=1)
```

## mean(average) -> sum of all elements / total no_of_elements

```python
np.mean(arr1)
```

```python
np.mean(arr2)
```

## var(variance) -> average of square difference from mean
    Mean = 2.5
    Differences = [-1.5, -0.5, 0.5, 1.5]
    Squared = [2.25, 0.25, 0.25, 2.25]
    Variance = (2.25 + 0.25 + 0.25 + 2.25) / 4 = 5 / 4 = 1.25

```python
np.var(arr1)
```

```python
np.var(arr2)
```

## std(standard deviation) -> square root of variance

```python
np.std(arr1)
```

```python
np.std(arr2)
```

```python
arr3=np.array(['s','u','m','a','n'])
```

```python
np.mean(arr3)
```

```python
np.var(arr3)
```

```python
np.std(arr3)
```

### np.sort()


```python
arr1=np.array([1,2,3,5,4])
arr2=np.array([[1000,100],[10,4]])
```

```python
np.sort(arr1)
```

```python
np.sort(arr2) # row wise performing the operation
```

```python
np.sort(arr2[0:,0:])
```

```python
arr5=np.array([1,10,20,30,5])
np.sort(arr5[:-1])
```

```python
arr6=np.array([[10,2,5],[20,200,1],[6,3,2]])
np.sort(arr6[:,::-1])
```

```python
np.sort(arr6[::1,::-1])
```
