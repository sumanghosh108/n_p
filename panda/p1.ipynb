```python
import pandas as pd
```

## Series creation:
syntax:
    pd.Series([values,.....])

```python
series=pd.Series([1,2,3,4,5.],index=['A','B','C','D','E'])
series
```

```python
series=pd.Series((1,2,3,4),index=('A','B','C','D'))
series
```

```python
series=pd.Series({'A':1,'B':2})
series
```

```python
pd.Series({'name':{'AA','BB'},'name1':['AA']})
```

```python
pd.Series({"name":['AA','BB'],"name1":['AA']})
```

## DataFrame creation:
Syntax: pd.DataFrame(dict)

```python
data=pd.DataFrame({'name':['rishi','mohon','suman'],'age':[22,22,22]})
data
```

```python
data=pd.DataFrame({'name':['rishi','mohon','suman','sunny','divya','sum','picku','tilak','scott','allen'],'age':[22,22,22,21,23,21,23,22,45,34]})
data
```

### second method to create a DataFrame:
    Syntax:
        pd.DataFrame(data,columns=[col_names])

```python
data1=pd.DataFrame([[1,2],[4,5]])
data1
```

```python
data1=pd.DataFrame([[1,2],[4,5]],columns=['col1','col2'])
data1
```
