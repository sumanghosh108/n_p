## isna() -> is a build-in method used to get null values from a dataset
    Syntax:
        df.isna() / df.isnull()
    -> returns boolean values
    -> returns True if the values are NaN else False

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isnull()
```

## dropna() -> it is a build-in method which is uesd to remove rows if it is having NaN(null value)
    -> if any column is having NaN it removes/deletes entire row by default
syntax:
    df.dropna()
    df.dropna(subset=[column1,column2,.....])
    df.dropna(subset=[column1,column2,.....],how='all')


```python
df
```

```python
df.dropna()
```

```python
df.dropna(subset=['col1']) # returns new dataframe
```

```python
df=df.dropna(subset=['col1'])
df
```

## how to update the dropna records by default
- df.dropna(inplace=True)

```python
df=pd.DataFrame({'col1':[1,2,3,np.nan,5,6],'col2':[11,12,None,None,100,20]})
df.dropna(inplace=True)
df
```

```python
df=pd.DataFrame({'col1':[1,2,3,np.nan,5,6],'col2':[11,12,None,None,100,20]})
df.dropna(subset=['col1','col2'],how='all')
```
